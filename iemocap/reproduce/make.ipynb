方針: セッションごとに分けたdataframeを作り，1セッションだけテスト，それ以外は訓練にする

```python
import pandas as pd
```

```python
df_all = pd.read_csv('iemocap_all.csv')
```

```python
df_all
```

```python
df_all['file'] = df_all['file'].map(lambda x: x[11:])
```

```python
df_all
```

```python
dfs_session = [pd.DataFrame(columns=df_all.columns) for i in range(5)]
```

```python
ses_dict = {1: 'Ses01', 2 : 'Ses02', 3 : 'Ses03', 4 : 'Ses04', 5 : 'Ses05'}
```

```python
keys = list(ses_dict.keys())
```

```python
vals = list(ses_dict.values())
```

```python
dfs_train = [pd.DataFrame(columns=df_all.columns) for i in range(5)]
```

```python
for i, val in enumerate(vals):
    for row in range(len(df_all)):
        if val in df_all['file'][row]:
            dfs_session[i] = dfs_session[i].append(df_all.iloc[row])
```

```python
dfs_session[0]
```

```python
for i in range(5):
    for j in range(5):
        if j is not i:
            dfs_train[i] = pd.concat([dfs_train[i], dfs_session[j]])
```

```python
dfs_train[2]
```

```python
for i in range(5):
    dfs_train[i].to_csv('iemocap_' + str(i + 1) + '.train.csv')
    dfs_session[i].to_csv('iemocap_' + str(i + 1) + '.test.csv')
```
